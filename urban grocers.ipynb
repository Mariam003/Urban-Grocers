```python
#change the date type of all the datasets
sales_df['Date'] = pd.to_datetime(sales_df['Date']) 
inventory_df['Date'] = pd.to_datetime(inventory_df['Date'])
workforce_df['Date'] = pd.to_datetime(workforce_df['Date'])
events_df['Date'] = pd.to_datetime(events_df['Date'])   
```

```python
#select numeric columns
sales_numeric_cols = sales_df.select_dtypes(include='number').columns
inventory_numeric_cols = inventory_df.select_dtypes(include='number').columns
workforce_numeric_cols = workforce_df.select_dtypes(include='number').columns
events_numeric_cols = events_df.select_dtypes(include='number').columns
```

```python
#aggregate numeric data by date
sales_daily = sales_df.groupby('Date')[sales_numeric_cols].mean().reset_index()
inventory_daily = inventory_df.groupby('Date')[inventory_numeric_cols].mean().reset_index()
workforce_daily = workforce_df.groupby('Date')[workforce_numeric_cols].mean().reset_index()   
events_daily = events_df.groupby('Date')[events_numeric_cols].mean().reset_index()  
```

```python
sales_daily.tail(3)
```

```python
print(sales_daily['Date'].nunique())
print(len(sales_daily))
```

```python
#merge
grocers_df = sales_daily \
    .merge(inventory_daily, on='Date', how='outer') \
    .merge(workforce_daily, on='Date', how='outer') \
    .merge(events_daily, on='Date', how='outer')
```

```python
grocers_df.head(3)
grocers_df.info()
```

```python
#extract day of week
grocers_df['day_of_week'] = grocers_df['Date'].dt.dayofweek
grocers_df['month'] = grocers_df['Date'].dt.month
grocers_df['week_of_year'] = grocers_df['Date'].dt.isocalendar().week
grocers_df['year'] = grocers_df['Date'].dt.year
```

```python
#forecasting
#import libraries
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
grocers_df.head(3)
```

```python
#flag weekends as sales behave differently on weekends
#where 1 is weekend and 0 is weekeday
grocers_df['weekend'] = grocers_df['day_of_week'].isin([5,6]).astype(int)
```

```python
grocers_df.head(3)
```

```python
#change column names
grocers_df.rename(columns = {
    'day_of_week': 'Day_Of_Week',
    'month': 'Month',
    'week_of_year': 'Weeks_Of_The_Year',
    'weekend': 'Weekend',
}, inplace = True)


```

```python
print(grocers_df)
```

```python
#forecasting sales
features = [
    'Temperature_C',
    'Opening_Inventory',
    'Inventory_Received',
    'Wastage_Units',
    'Hours_Worked',
    'Daily_Sales_Target',
    'Day_Of_Week',
    'Month',
    'Weeks_Of_The_Year',
    'Weekend',
    'year' 
]

X = grocers_df[features]
y = grocers_df['Sales_Amount']
```

```python
#split and train into sets
#80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.2, shuffle = False
)    
```

```python
# Initialize the model
xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
```

```python
# Train the model on the training set
xgb_model.fit(X_train, y_train)
```

```python
#predict on the test set
y_pred = xgb_model.predict(X_test)
```

```python
#calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
# Print results
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

```python
import matplotlib.pyplot as plt

# Plot actual vs predicted sales for the first 100 test samples
plt.figure(figsize=(12, 6))
plt.plot(y_test.values[:100], label='Actual Sales', marker='o')
plt.plot(y_pred[:100], label='Predicted Sales', marker='x')
plt.xlabel('Days')
plt.ylabel('Sales Amount')
plt.title('Actual vs Predicted Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a DataFrame with actual and predicted values
results_df = pd.DataFrame({
    'Actual_Sales': y_test.values,
    'Predicted_Sales': y_pred
})
```

```python
# Reset index for time series alignment
results_df.reset_index(drop=True, inplace=True)
```

```python
# Export to CSV
results_df.to_csv('sales_predictions.csv', index=False)

print("CSV file saved as 'sales_predictions.csv'")
```
